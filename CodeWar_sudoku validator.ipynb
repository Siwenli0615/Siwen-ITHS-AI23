```python

def validate_sudoku(board):
  check_control = [x for x in range(1,10)]
  row = []
  column = []
  #check rows
  for i in range(9):
    row = board[i]
    row = sorted(row)
    if row == check_control:
      continue
    else:
      return False
  #check columns
  for k in range(9):
    for j in board:
      column.append(j[k])

    column = sorted(column)
    if column == check_control:
      column = []
      return True
    else:
      return False
      break

validate_sudoku(board)
```
